# Model

```python
#Creating a pipeline and fitting standard scalar to the values in labels

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

my_pipeline = Pipeline([('std_scalar',StandardScaler())])
```

```python
def print_rmse(name,labels,value_predictions):
    mse = mean_squared_error(labels,value_predictions)
    rmse = np.sqrt(mse)
    print(f'RMSE for {name} = {rmse}')
```

```python
#Fitting the features in the pipeline
feat_pipe = my_pipeline.fit_transform(features)
# feat pipe is the features after going through the pipeline.. Returned as a numpy array
# print(np.array(feat_pipe[4]))
# labels
feat_pipe
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

#Creating an object of class
model = RandomForestRegressor()
#fitting orignal features and labels of the training dataset
model.fit(feat_pipe,labels)

#Predicting the labels with features of the training set as input
value_predictions = model.predict(feat_pipe)
```

```python
print_rmse("Random Forest Regressor", labels,value_predictions)
```

```python
model.score(feat_pipe,labels)
```

```python
from sklearn.metrics import mean_squared_error
feat_test = test_set.drop("value_eur",axis = 1)
label_test = test_set["value_eur"].copy()
#X_testprep is the features of the test set returned as numpy array
X_test_prepared = my_pipeline.fit_transform(feat_test)

value_predictions_test = model.predict(X_test_prepared)
```

```python
print_rmse("Random Forest Regressor test", label_test,value_predictions_test)
```

```python
model.score(X_test_prepared,label_test)
```
